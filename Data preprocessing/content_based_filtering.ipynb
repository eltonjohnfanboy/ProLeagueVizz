```python
relation_df = pd.DataFrame(index = mastery_rel['Champion'], columns = mastery_rel['Champion'])
relation_df = relation_df.fillna(0)

for x in mastery_rel['Champion']:
    con_list = []
    for rel in ['Rel1', 'Rel2', 'Rel3', 'Rel4', 'Rel5']:
        con_list.append(mastery_rel[mastery_rel['Champion'] == x][rel])

    for c in con_list:
        if not np.isnan(c.item()) and c.item() != 0:
            relation_df[x][c.item()] = 1
            relation_df[c.item()][x] = 1
```

```python
mapped_index = relation_df.index.map(map_idName.set_index('ID')['Name'])
mapped_columns = relation_df.columns.map(map_idName.set_index('ID')['Name'])

relation_df.index = mapped_index
relation_df.columns = mapped_columns
np.fill_diagonal(relation_df.values, 1)

relation_df = relation_df.sort_index(axis=0).sort_index(axis=1)
```

```python
relation_df
```

```python
# Remove columns/index that are not common

a = set(relation_df.index)
b = set(nature_similarityDf.index)
list_to_remove = list(b-a)

nature_similarityDf = nature_similarityDf.drop(index=list_to_remove, columns=list_to_remove)
similarity_df = similarity_df.drop(index=list_to_remove, columns=list_to_remove)
```

```python
# Define weights for each DataFrame
weights = [0.45, 0.25, 0.3]

# Calculate the weighted sum of the DataFrames
final_similarity_df = (relation_df * weights[0]) + (nature_similarityDf * weights[1]) + (similarity_df * weights[2])
```
