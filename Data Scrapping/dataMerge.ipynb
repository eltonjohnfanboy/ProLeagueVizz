```python
import pandas as pd
from fuzzywuzzy import fuzz
```

```python
df1 = pd.read_csv('C:/Users/adars/OneDrive/Escritorio/ProjecteLolShiny/Data/PlayersGolggStats.csv')
df2 = pd.read_csv('C:/Users/adars/OneDrive/Escritorio/ProjecteLolShiny/Data/PlayerStatsGeneral.csv')
```

```python
# Function to find best match between two events
# Function to find best match between two events
def find_best_match(event, event_list, threshold=50):
    best_score = 0
    best_match = None
    event_has_playoffs = "Playoffs" in event
    for event_in_list in event_list:
        score = fuzz.token_set_ratio(event, event_in_list)
        match_has_playoffs = "Playoffs" in event_in_list
        if score > best_score and score >= threshold and (event_has_playoffs or not match_has_playoffs):
            best_score = score
            best_match = event_in_list
    return best_match if best_score >= threshold else None

# Matching events between df1 and df2
matches = {}
for event in df1['Event']:
    match = find_best_match(event, df2['Event'])
    matches[event] = match

# Create a new column in df1 to store the matched events
df1['matched_event'] = df1['Event'].map(matches)
```
